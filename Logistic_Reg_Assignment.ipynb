Fixing Outliers

here will replace the lower one with min IQR

and higher one with Q3+1.5*IQR

```
def remove_outlier(df_bank,col_name):
  Q1=df_bank[col_name].quantile(0.25)
  Q3=df_bank[col_name].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=df_bank[col_name].min()
  outer_bound=Q3+1.5*Q1
  print("colomn",col_name,"lower and outer bound values are :",lower_bound,outer_bound)
  df_outer=df_bank.loc[(df_bank[col_name]>lower_bound) & (df_bank[col_name]<outer_bound)]
  return df_outer
```

```
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 
    lower_bound = df_in[col_name].min()
    upper_bound = q3+1.5*iqr
    print('Column',col_name,'IQR lower bound and upper bound are', lower_bound, 'and', upper_bound, 'respectively')
    df_out = df_in.loc[(df_in[col_name] > lower_bound) & (df_in[col_name] < upper_bound)]
    return df_out
```

```
bank_cat=bank[['job', 'marital','default', 'education', 'loan', 'housing', 'contact', 'month', 'day_of_week', 'poutcome', 'y']]
```

```
bank_cont=bank.drop(['job', 'marital','default', 'education', 'loan', 'housing', 'contact', 'month', 'day_of_week', 'poutcome', 'y'],axis=1)
```

```
from sklearn.preprocessing import LabelEncoder
bank_cat=bank_cat.apply(LabelEncoder().fit_transform)
```

```
bank_final=pd.concat([bank_cat,bank_cont],axis=1)
```

```
bank_final.head()
```

```
x=bank_final.loc[:,bank_final.columns!="y"]
```

```
print(x.shape)
x.head()
```

```
y=bank_final.loc[:,bank_final.columns=="y"]
y.head()
```

```
from sklearn.model_selection import train_test_split
def split(x,y):
  return train_test_split(x,y,test_size=0.25,random_state=1)
```

```
xtrain,xtest,ytrain,ytest=split(x,y)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```
bank_final.head()
```

```
bank_final.info()
```

```
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
```

```
ytrain_pred=logreg.predict(xtrain)
ytest_pred=logreg.predict(xtest)
```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(ytest,ytest_pred)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix=pd.DataFrame(confusion_matrix(ytest,ytest_pred))
print(confusion_matrix)
```

```
confusion_matrix.index=["actual_No_Deposit","actual_Deposit"]
confusion_matrix.columns=["predicted_No_Deposit","predicted_Deposit"]
confusion_matrix
```

```
y_test_pred1=np.where(logreg.predict_proba(xtest)[:,1]>0.75,1,0)
accuracy_score(ytest,y_test_pred1)
```
