```
stocks_df=pd.read_csv("https://raw.githubusercontent.com/benjaminwilson/python-hierarchical-clustering-exercises/master/datasets/company-stock-movements-2010-2015-incl.csv",index_col=0)
```

```
stocks_df.head()
```

```
stocks_df.index
```

```
companies=stocks_df.index
```

```
companies
```

```
movements=stocks_df.values
```

```
movements
```

```
from sklearn.preprocessing import normalize
```

```
from scipy.cluster.hierarchy import linkage,dendrogram
```

```
import matplotlib.pyplot as plt
```

```
movements_normalize=normalize(movements)
```

```
merge=linkage(movements_normalize,method="complete")
```

```
plt.figure(figsize=(15,7))
dendrogram(merge,
    labels=companies,
    leaf_rotation=90.,
    leaf_font_size=10)
plt.show()
```
